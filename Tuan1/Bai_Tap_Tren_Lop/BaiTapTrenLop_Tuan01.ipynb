```python
#Bài 13: 
def tron_mang(a, b):
    ket_qua = []
    do_dai_max = max(len(a), len(b))
    
    for i in range(do_dai_max):
        gia_tri_a = a[i] if i < len(a) else 0
        gia_tri_b = b[i] if i < len(b) else 0
        ket_qua.append(gia_tri_a + gia_tri_b)
        
    return ket_qua

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]
print(tron_mang(a, b))
```

```python
#Bài 14: Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau
import random
import math
def tao_mang_ngau_nhien(m, n):
    return [[random.randint(1, 20) for _ in range(n)] for _ in range(m)]

def xuat_dong_k(a, k):
    return a[k] if 0 <= k < len(a) else []

def xuat_cot_k(a, k):
    return [row[k] for row in a] if 0 <= k < len(a[0]) else []

def tim_dong_tong_max_duoi_45(a):
    hop_le = [(i, sum(row)) for i, row in enumerate(a) if sum(row) <= 45]
    return max(hop_le, key=lambda x: x[1])[0] if hop_le else -1

def tim_cot_tich_min(a):
    n_cols = len(a[0])
    tich_cac_cot = [math.prod(a[i][j] for i in range(len(a))) for j in range(n_cols)]
    return tich_cac_cot.index(min(tich_cac_cot))

def xuat_dong_chan_cot_le(a):
    return [a[i][j] for i in range(len(a)) if i % 2 == 0 
            for j in range(len(a[0])) if j % 2 != 0]

def tbc_chan_dong_le(a):
    so_chan = [a[i][j] for i in range(len(a)) if i % 2 != 0 
               for j in range(len(a[0])) if a[i][j] % 2 == 0]
    return sum(so_chan) / len(so_chan) if so_chan else 0

def tbc_bien(a):
    m, n = len(a), len(a[0])
    bien = [a[i][j] for i in range(m) for j in range(n) 
            if i == 0 or i == m - 1 or j == 0 or j == n - 1]
    return sum(bien) / len(bien) if bien else 0

def tbt_khong_bien(a):
    m, n = len(a), len(a[0])
    trong = [a[i][j] for i in range(1, m - 1) for j in range(1, n - 1)]
    if not trong: return 0
    return math.prod(trong) ** (1 / len(trong))
#Tạo mảng
m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))
a = tao_mang_ngau_nhien(m, n)

print("=== MẢNG A===")
for row in a:
    print(row)

print("\n KẾT QUẢ CÁC CÂU ")
print(f"b. Các phần tử thuộc dòng 1: {xuat_dong_k(a, 1)}")
print(f"c. Các phần tử thuộc cột 2: {xuat_cot_k(a, 2)}")
print(f"d. Chỉ số dòng có tổng lớn nhất (<= 45): {tim_dong_tong_max_duoi_45(a)}")
print(f"e. Chỉ số cột có tích nhỏ nhất: {tim_cot_tich_min(a)}")
print(f"f. Các phần tử thuộc dòng chẵn và cột lẻ: {xuat_dong_chan_cot_le(a)}")
print(f"g. TBC các phần tử chẵn thuộc dòng lẻ: {tbc_chan_dong_le(a):.2f}")
print(f"h. TBC các phần tử thuộc biên: {tbc_bien(a):.2f}")
print(f"i. Trung bình tích các phần tử không thuộc biên: {tbt_khong_bien(a):.2f}")
```
