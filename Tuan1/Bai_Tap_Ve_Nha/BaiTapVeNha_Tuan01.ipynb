```python
#Bài 1
def tron_hai_mang(a, b):
    ket_qua = []
    min_len = min(len(a), len(b))
    for i in range(min_len):
        nho_nhat = min(a[i], b[i])
        ket_qua.append(nho_nhat)
    if len(a) > min_len:
        ket_qua.extend(a[min_len:])
    elif len(b) > min_len:
        ket_qua.extend(b[min_len:])
    return ket_qua

mang_a = [3, 9, 1, 4]
mang_b = [2, 7, 4, 3, 2, 8]

res = tron_hai_mang(mang_a, mang_b)
print(f"Mảng kết quả: {res}")
```

```python
#Bài 2
import math

# a. Tính tổng tam giác trên (bao gồm đường chéo phụ)
def tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

# b. Chuyển phần tử âm thành trị tuyệt đối
def am_sang_tri_tuyet_doi(a, m, n):
    for i in range(m):
        for j in range(n):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])

# c. Thay phần tử chẵn bằng x
def thay_chan_bang_x(a, m, n, x):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x

# d. Kiểm tra ma trận toàn chẵn
def kiem_tra_toan_chan(a, m, n):
    for row in a:
        for val in row:
            if val % 2 != 0:
                return False
    return True

# e. Kiểm tra ma trận đối xứng (a[i][j] == a[j][i])
def kiem_tra_doi_xung(a, n):
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

# f. Kiểm tra đường chéo chính tăng dần
def cheo_chinh_tang_dan(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

# g. Xuất tam giác dưới (bao gồm đường chéo phụ)
def xuat_tam_giac_duoi_cheo_phu(a, n):
    print("Các phần tử tam giác dưới đường chéo phụ:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()

# h. Kiểm tra đường chéo phụ giảm dần
def cheo_phu_giam_dan(a, n):

    for i in range(n - 1):
        if a[i][n-1-i] <= a[i+1][n-1-(i+1)]:
            return False
    return True
    # Giả sử ta có ma trận vuông cấp 3 (n=3)
ma_tran = [
    [3, 5, 1],
    [-2, 8, 4],
    [7, 0, 9]
]
n = 3
x = 99

print("--- MA TRẬN BAN ĐẦU ---")
for row in ma_tran: print(row)

# a. Tổng tam giác trên đường chéo phụ (3, 5, 1, -2, 8, 7)
print(f"\na. Tổng tam giác trên chéo phụ: {tong_tam_giac_tren_cheo_phu(ma_tran, n)}")

# d. Kiểm tra toàn chẵn
print(f"d. Ma trận toàn chẵn? {kiem_tra_toan_chan(ma_tran, n, n)}")

# e. Kiểm tra đối xứng
print(f"e. Ma trận đối xứng? {kiem_tra_doi_xung(ma_tran, n)}")

# f. Kiểm tra chéo chính tăng dần (3 < 8 < 9)
print(f"f. Chéo chính tăng dần? {cheo_chinh_tang_dan(ma_tran, n)}")

# h. Kiểm tra chéo phụ giảm dần (1 > 8 > 7) -> Sai, 8 không nhỏ hơn 7
print(f"h. Chéo phụ giảm dần? {cheo_phu_giam_dan(ma_tran, n)}")

# g. Xuất tam giác dưới đường chéo phụ
print("\ng. Tam giác dưới chéo phụ:")
xuat_tam_giac_duoi_cheo_phu(ma_tran, n)

# b & c. Thay đổi giá trị (Trị tuyệt đối & Thay số chẵn)
am_sang_tri_tuyet_doi(ma_tran, n, n)
thay_chan_bang_x(ma_tran, n, n, x)

print("\n--- MA TRẬN SAU KHI SỬA (câu b & c) ---")
for row in ma_tran: print(row)
```
